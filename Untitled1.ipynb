```python
import matplotlib.pyplot as plt
import numpy as np

# Años de análisis
anios = np.arange(2015, 2025)

# --- Datos de Exportaciones por Sector y País (en miles de millones de USD) ---
# Datos representativos basados en tendencias de mercado para comparación.

# Sector Minero
mineria = {
    'Perú': [19.1, 22.1, 27.2, 30.3, 28.4, 28.3, 39.6, 39.5, 42.8, 47.7],
    'Chile': [33.5, 36.8, 41.5, 45.9, 42.1, 41.2, 53.6, 56.1, 54.2, 57.8],
    'Colombia': [1.5, 1.8, 2.1, 2.3, 2.2, 2.0, 2.5, 2.8, 2.9, 3.1],
    'Brasil': [45.1, 46.5, 50.8, 55.4, 52.3, 50.5, 65.2, 68.4, 69.8, 71.5]
}

# Sector Agropecuario
agro = {
    'Perú': [5.2, 5.5, 6.4, 7.1, 7.4, 8.0, 9.2, 9.8, 10.5, 12.8],
    'Chile': [10.5, 11.2, 12.0, 12.5, 13.0, 13.5, 15.0, 16.2, 17.0, 18.5],
    'Colombia': [7.8, 8.0, 8.5, 9.1, 9.5, 10.0, 10.8, 11.5, 12.0, 13.2],
    'Brasil': [88.5, 91.0, 96.5, 101.0, 105.0, 112.0, 125.0, 138.0, 145.0, 155.0]
}

# Sector Pesquero
pesca = {
    'Perú': [1.8, 2.0, 2.9, 3.4, 2.7, 2.6, 4.0, 3.9, 3.6, 3.6],
    'Chile': [5.5, 5.8, 6.2, 6.5, 6.3, 6.0, 6.8, 7.0, 7.1, 7.3],
    'Colombia': [0.15, 0.18, 0.2, 0.22, 0.21, 0.2, 0.25, 0.28, 0.3, 0.32],
    'Brasil': [0.5, 0.6, 0.65, 0.7, 0.68, 0.65, 0.75, 0.8, 0.85, 0.9]
}

# --- Creación de los Subgráficos ---
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(14, 20))
plt.subplots_adjust(hspace=0.4)

def crear_grafico_lineas(ejes, datos, titulo):
    for pais, valores in datos.items():
        ejes.plot(anios, valores, 'o-', label=pais, linewidth=2)
    ejes.set_title(titulo, fontsize=16, fontweight='bold')
    ejes.set_ylabel('Valor (Miles de Millones de USD)', fontsize=12)
    ejes.grid(True, linestyle='--', alpha=0.6)
    ejes.legend(title='Países', loc='upper left')

# Gráfico del Sector Minero
crear_grafico_lineas(ax1, mineria, 'Comparación de Exportaciones Mineras (2015-2024)')

# Gráfico del Sector Agropecuario
crear_grafico_lineas(ax2, agro, 'Comparación de Exportaciones Agropecuarias (2015-2024)')

# Gráfico del Sector Pesquero
crear_grafico_lineas(ax3, pesca, 'Comparación de Exportaciones Pesqueras (2015-2024)')
ax3.set_xlabel('Años', fontsize=12)

# Muestra el gráfico
plt.show()
```

```python
# ============================================
# Exportaciones Peruanas Detalladas - Google Colab
# 3 productos por sector (Minería, Agro, Pesca)
# Con origen (región) y destino (país)
# ============================================

import pandas as pd
import matplotlib.pyplot as plt

# 1. Datos detallados (ficticios pero realistas)
data = {
    'Sector': [
        'Minería','Minería','Minería',
        'Agro','Agro','Agro',
        'Pesca','Pesca','Pesca'
    ],
    'Producto': [
        'Cobre','Oro','Zinc',
        'Café','Espárragos','Quinua',
        'Harina de Pescado','Pota Congelada','Conserva de Atún'
    ],
    'Origen (Región)': [
        'Arequipa','La Libertad','Pasco',
        'Junín','Ica','Puno',
        'Chimbote','Piura','Callao'
    ],
    'Destino (País)': [
        'China','Suiza','Corea del Sur',
        'Alemania','EE.UU.','Canadá',
        'Japón','España','Chile'
    ],
    'Valor Exportado (Millones USD)': [
        15000,8000,5000,
        1200,950,300,
        2000,800,400
    ]
}

df = pd.DataFrame(data)

# 2. Tabla detallada completa
print("=== Tabla Detallada de Exportaciones (3 Minerías, 3 Agro, 3 Pesca) ===")
display(df)

# 3. Totales por sector
totales_sector = df.groupby('Sector')['Valor Exportado (Millones USD)'].sum().reset_index()
print("\n=== Totales por Sector ===")
display(totales_sector)

# 4. Totales por país destino
totales_pais = df.groupby('Destino (País)')['Valor Exportado (Millones USD)'].sum().reset_index()
print("\n=== Totales por País de Destino ===")
display(totales_pais)

# 5. Gráfico de barras apiladas por sector y producto
plt.figure(figsize=(12,8))

# Colores distintos para cada producto
colors = ['#d9534f','#f0ad4e','#5bc0de','#5cb85c','#428bca','#e67e22','#8e44ad','#2ecc71','#3498db']

# Inicializar los acumuladores por sector
sectores = df['Sector'].unique()
bottoms = {sector:0 for sector in sectores}

# Dibujar barras apiladas
for idx,row in df.iterrows():
    plt.bar(row['Sector'],
            row['Valor Exportado (Millones USD)'],
            bottom=bottoms[row['Sector']],
            color=colors[idx],
            label=row['Producto'] if row['Producto'] not in plt.gca().get_legend_handles_labels()[1] else "")
    bottoms[row['Sector']]+=row['Valor Exportado (Millones USD)']

plt.title('Exportaciones Peruanas Detalladas por Sector y Producto (Millones USD)',fontsize=14)
plt.ylabel('Millones USD')
plt.xlabel('Sector')
plt.legend(title='Producto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 6. Extra: Resumen general para que quede más espectacular
print("\n=== Resumen General ===")
print(f"Total exportado en los 3 sectores: {df['Valor Exportado (Millones USD)'].sum()} Millones USD")
print("Principales países de destino: ", ', '.join(df['Destino (País)'].unique()))
print("Principales regiones de origen: ", ', '.join(df['Origen (Región)'].unique()))
```

```python
# --- Exportaciones detalladas de Perú con porcentajes y mapa interactivo ---
import pandas as pd
import matplotlib.pyplot as plt
import folium

# Datos: 3 productos de minería, 3 de agro, 3 de pesca
data = {
    'Sector': ['Minería', 'Minería', 'Minería',
               'Agro', 'Agro', 'Agro',
               'Pesca', 'Pesca', 'Pesca'],
    'Producto': ['Cobre', 'Oro', 'Zinc',
                 'Café', 'Espárragos', 'Palta',
                 'Harina de pescado', 'Langostinos', 'Conchas de abanico'],
    'Región (Perú)': ['Arequipa', 'La Libertad', 'Cerro de Pasco',
                      'Junín', 'Ica', 'Lambayeque',
                      'Chimbote', 'Tumbes', 'Pisco'],
    'Destino (País)': ['China', 'Suiza', 'Corea del Sur',
                       'Alemania', 'Estados Unidos', 'Países Bajos',
                       'China', 'España', 'Francia'],
    'Valor (Millones USD)': [5000, 4200, 1500, 800, 600, 700, 1200, 300, 150]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Calcular % dentro de cada sector para cada país destino
grouped = df.groupby(['Sector','Destino (País)'])['Valor (Millones USD)'].sum().unstack().fillna(0)
percent = grouped.div(grouped.sum(axis=1), axis=0)*100

# Crear DataFrame de porcentajes fusionado con datos
df_percent = df.copy()
df_percent['% País dentro del sector'] = df.apply(
    lambda row: (row['Valor (Millones USD)'] /
                 grouped.loc[row['Sector'], row['Destino (País)']]) *
                 (grouped.loc[row['Sector'], row['Destino (País)']] /
                  grouped.loc[row['Sector']].sum())*100, axis=1)

print("=== Tabla detallada con porcentajes ===")
display(df_percent)

# Gráfico apilado en porcentajes con etiquetas
fig, ax = plt.subplots(figsize=(12,7))
percent.plot(kind='bar', stacked=True, ax=ax, cmap='tab20')
plt.title("Exportaciones Peruanas por Sector y País Destino (%)", fontsize=16)
plt.xlabel("Sector", fontsize=12)
plt.ylabel("Porcentaje (%)", fontsize=12)
plt.legend(title="País Destino", bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()

# --- Mapa interactivo ---
# Coordenadas simples (Países destino y Perú)
coords = {
    'Perú': [-9.19, -75.02],
    'China': [35.86, 104.19],
    'Suiza': [46.81, 8.22],
    'Corea del Sur': [35.91, 127.77],
    'Alemania': [51.16, 10.45],
    'Estados Unidos': [37.09, -95.71],
    'Países Bajos': [52.13, 5.29],
    'España': [40.46, -3.75],
    'Francia': [46.22, 2.21]
}

m = folium.Map(location=[-9.19, -75.02], zoom_start=2)

# Marcador de Perú
folium.Marker(coords['Perú'], popup='Perú (Origen)', icon=folium.Icon(color='green')).add_to(m)

# Marcadores de países destino
for idx, row in df.iterrows():
    destino = row['Destino (País)']
    folium.Marker(
        coords[destino],
        popup=f"{row['Producto']} desde {row['Región (Perú)']} a {destino}: "
              f"{row['Valor (Millones USD)']}M USD",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(m)

m
```
